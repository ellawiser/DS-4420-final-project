```python
import pandas as pd
```

```python
df = pd.read_csv("Data/cancer_image_csv_data/breast_cancer_images.csv")
df.head()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_gen = datagen.flow_from_dataframe(
    df,
    x_col="image_path",
    y_col="classification",
    target_size=(224,224),
    class_mode="categorical",
    subset="training",
    batch_size=32,
    shuffle=True
)

val_gen = datagen.flow_from_dataframe(
    df,
    x_col="image_path",
    y_col="classification",
    target_size=(224,224),
    class_mode="categorical",
    subset="validation",
    batch_size=32,
    shuffle=False
)
```
